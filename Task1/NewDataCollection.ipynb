```python
from newsapi.newsapi_client import NewsApiClient
import pandas as pd
import json
from dotenv import load_dotenv
load_dotenv(verbose=True)
import os
APIKey = os.getenv("APIKey")

api = NewsApiClient(api_key=APIKey)
```

```python
all_articles = api.get_everything(    sources='google-news',
                                      from_param='2021-03-11',
                                      to='2021-03-12',
                                      language='en')
```

```python
all_articles
```

### Get the news content from March 11, 2021 to March 12, 2021

```python
for i in range(len(all_articles['articles'])):
    date = all_articles['articles'][i]['publishedAt'],
    title = all_articles['articles'][i]['title'],
    full_article = all_articles['articles'][i]['content'],
    author = all_articles['articles'][i]['author']
    print(f'Title is {title} made by {author} full article: {full_article} ')
```

### Convert to JSON

```python
news_json = []
for i in range(len(all_articles['articles'])):
    news_json.append({
                "date": all_articles['articles'][i]['publishedAt'],
                "title": all_articles['articles'][i]['title'],
                "full_article": all_articles['articles'][i]['content'],
                "author": all_articles['articles'][i]['author']
            })
```
